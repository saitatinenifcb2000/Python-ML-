# Handle Missing Values

```python
# NaN, NaT, None
```

```python
# Creating Lists
names = ['A', "B", 'C', 'D', np.NaN]
phy = [12, 13, np.NaN, 15, np.NaN]
maths = [45, 4, 1, np.NaN,np.NaN]
pet = [np.NaN]*5
pet
```

```python
# Creating a DF
mdf = pd.DataFrame({"Name" : names, "Physics": phy, "Maths" : maths, "PET": pet})
mdf
```

```python
# Dropping nulls from rows
mdf.dropna(how = "any", axis = 0) # Axis = 0 Means Rows and 1 Means Columns
```

```python
# Dropping nulls from columns
mdf.dropna(how = "any", axis = 1) # Axis = 0 Means Rows and 1 Means Columns
```

```python
mdf
```

```python
# Dropping nulls from rows if all the values are null
mdf.dropna(how = "all", axis = 0) # Axis = 0 Means Rows and 1 Means Columns
```

```python
# Dropping nulls from columns if all the values are null
mdf.dropna(how = "all", axis = 1) # Axis = 0 Means Rows and 1 Means Columns , inplace = True
```

# Replace Nulls

```python
mdf
```

```python
# Replacing Nulls with a Constant
mdf.fillna(10)
```

```python
# Replacing Nulls within a Column with a value
mdf['Name'].fillna('E')
```

```python
# Replacing Nulls within a numerical Column with a value
mdf['Physics'].fillna(round(mdf['Physics'].mean(), 2), inplace = True)
```
