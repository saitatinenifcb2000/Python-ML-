```python
# Importing required Libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import random
import names
```

```python
# Importing Seaborn with Styles
import seaborn as sns
sns.set_style('darkgrid') # white, dark, whitegrid, darkgrid, ticks
```

```python
# Loading a Table
df = pd.read_excel("Sample - Superstore.xlsx")
df.head(2)
```

```python
# Checking number of unique vals
df['City'].nunique()
```

# Top 10 Cities by Profit

```python
top10cities = df.groupby('City')['Profit'].sum().sort_values(ascending = False)[:10]
top10cities = top10cities.sort_index()
top10cities
```

```python
city = list(top10cities.index)
vals = list(top10cities.values)
city
```

```python
vals
```

```python
top10cities = pd.DataFrame({"City": city, "Values": vals})
top10cities
```

# Line Chart

```python
# Plotting a Line Chart
plt.figure(figsize = (15, 5))
plt.title("Top 10 Cities by Profit")
sns.lineplot(x = 'City', y = 'Values', data = top10cities);
```

```python
top10 = df.groupby('City')['Profit'].sum().sort_values(ascending = False)[:10]
top10
```

# Bar Chart

```python
# Plotting a Bar Chart
# sns.color_palette('CMRmap_r')
plt.figure(figsize = (15, 5))
plt.title("Top 10 Cities by Profit")
sns.barplot(x = 'City', y = 'Values', data = top10cities);
```

```python
# Plotting an Horizontal Bar Chart
plt.figure(figsize = (15, 5))
plt.title("Top 10 Cities by Profit")
sns.barplot(y = 'City', x = 'Values', data = top10cities);
```

# When to use what chart?
* 1. Int - Bar chart
* 2. Frequency, Contingency Table- Histogram
* 3. Float - Scatterplot, bubbleplot, categorical plot
* 4. Outliers - Boxplot, violinplot
* 5. Interaction b/w two numerical cols - scatterplot, jointplot, pairplot
* 6. Correlation - Heatmap
* 7. Trend - Line chart, area chart
* 8. % of Total - Pie chart
* 9. Distribution - Line chart, scatter plots, catplot, jointplot
