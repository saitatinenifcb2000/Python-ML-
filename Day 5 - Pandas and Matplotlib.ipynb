# Join and Merge

```python
# Joins - Uses Index to merge dataframes
# Merge - Uses columns to merge dataframes
```

```python
orders.columns
```

```python
df1 = orders.loc[:10,['Region', 'Sales', 'Profit']]
df1
```

```python
df2 = people.copy()
df2
```

```python
# Inner Join
pd.merge(df1, df2, on = 'Region', how = 'inner')
```

```python
# Left Join
pd.merge(df1, df2, on = 'Region', how = 'left')
```

```python
# Right Join
pd.merge(df1, df2, on = 'Region', how = 'right')
```

```python
# Right Join
pd.merge(df1, df2, on = 'Region', how = 'outer')
```

```python
# Creating a Copy of the existing Df and Modifying its column names
df3 = df2.copy()
df3.columns = ['Manager', 'Location']
```

```python
df3
```

```python
df1
```

```python
# Merging DF's based on diff column names
pd.merge(df1, df3, left_on = 'Region', right_on = 'Location', how = 'inner')
```

```python
# Column Prefixes
pd.merge(df1, df3, left_on = 'Region', right_on = 'Location', how = 'inner')
```
