```python
# SUBPLOT
# Figure = Entire Chart(all charts in both rows and columns) 
# Axes = Individual Charts

fig, ax = plt.subplots(nrows = 2, ncols = 2, figsize = (10, 10))

# Plotting Visuals
ax[0, 0].plot(a, 'g--')
ax[0, 1].plot(b, 'r^')
ax[1, 0].plot(a, b, 'm--o')
ax[1, 1].plot(b, a, 'k*')

# Adding Titles
ax[0, 0].set_title('Var A')
ax[0, 1].set_title('Var B')
ax[1, 0].set_title('A vs B')
ax[1, 1].set_title('B vs A')

# Adding Title to the Entire Figure
fig.suptitle('Building Subplots for the first time')

# Controlling Scales in Subplots
ax[0, 0].set_ylim(0, 15)
ax[0, 1].set_ylim(0, 150)

ax[1, 0].set_xlim(0, 15)
ax[1, 0].set_ylim(0, 150)

ax[1, 1].set_xlim(0, 150)
ax[1, 1].set_ylim(0, 15)

# Labels in Subplots
ax[0, 0].set_xlabel("Var A")
ax[0, 0].set_ylabel("Var A")

ax[0, 1].set_xlabel("Var B")
ax[0, 1].set_ylabel("Var B")

ax[1, 0].set_xlabel("Var A")
ax[1, 0].set_ylabel("Var B")

ax[1, 1].set_xlabel("Var B")
ax[1, 1].set_ylabel("Var A");
```

# Top 10 Sub-categories by sum of Profit

```python
# Creating Top 10 Sub Categories by Profit
top10cats = df.groupby('Sub-Category')['Profit'].sum().sort_values(ascending = False)[:10]
top10cats
```

```python
# Sorting based on Index
top10cats = top10cats.sort_index()
```

```python
top10cats
```

```python
# Building X axis
x = list(top10cats.index)
x
```

```python
# Building Y axis
y = list(top10cats.values)
y
```

# Line Chart

```python
plt.figure(figsize = (15, 7))
plt.title("Top 10 Sub Categories by Profit")
plt.plot(x, y, 'r--o')
plt.xlabel('Sub-Categories')
plt.ylabel("Profit");
```

# Bar Chart

```python
# Budiling a Vertical Bar Chart
plt.figure(figsize = (15, 5))
plt.title("Top 10 Sub Categories by Profit")
plt.bar(x, y, color = ['r', 'g', 'b', 'y', 'k']*2)
plt.xlabel('Sub-Categories')
plt.ylabel("Profit");
```

```python
sorted(x, reverse = True)
```

```python
# Budiling a Horizontal Bar Chart
plt.figure(figsize = (15, 5))
plt.title("Top 10 Sub Categories by Profit")
plt.barh(x, y, color = ['r', 'g', 'b', 'y', 'k']*2)
plt.xlabel('Profit')
plt.ylabel("Sub-Categories");
```

```python
# Budiling a Horizontal Bar Chart
plt.figure(figsize = (15, 5))
plt.title("Top 10 Sub Categories by Profit")
plt.barh(x, sorted(y, reverse=True), color = ['r', 'g', 'b', 'y', 'k']*2)
plt.xlabel('Profit')
plt.ylabel("Sub-Categories");
```
